```python
# 선수
#driver.execute_async_script("javascript:moveMainFrame('0410');")
#print( '데이터 진입' )
# 1초 대기 
#time.sleep(3)
#print( '추출 시작' )
# 데이터 추출
# 페이지 내의 선수기록 테이블 추출하기
GK = '#frm > table:nth-child(102)'
DF = '#frm > table:nth-child(428)'
MF = '#frm > table:nth-child(685)'
FW = '#frm > table:nth-child(940)'

position = [GK, DF, MF, FW]


table = driver.find_element(By.CSS_SELECTOR, GK)
# thead
thead = table.find_element(By.TAG_NAME, "thead")
tbody = table.find_element(By.TAG_NAME, "tbody")

col = thead.text.split(' ')
col[0] = col[0]+col[1]
del col[1]

table_rows = table.find_elements(By.TAG_NAME, 'tr')
datas = []
for table_row in table_rows[1:]:
    row_data = table_row.text.split(' ')
    if len(row_data) == 7:
        modified_name = "".join(row_data[0:2])  
        row_data[0:2] = [modified_name]
    if len(row_data) == 8:
        modified_name = "".join(row_data[0:3]) 
        row_data[0:3] = [modified_name]
    datas.append(row_data) 
#print(datas)
gk1 = pd.DataFrame(datas, columns=col)
gk1
# dfs= pd.concat([dfs, df])
# dfs.head(5), dfs.tail(5)

# 홈페이지 이동
#driver.get(target)
```

```python
gk1.to_excel('GK_2023.xlsx', index=False)
```
